# BAR PLOT

```python
import matplotlib.pyplot as plt
import numpy as np
a = ["Dev", "Devasya", "Khushi", "Purva"]
b = [9.08, 7.12, 9.47, 8.37]
c = [9.20, 8.01, 9.67, 8.99]

width = 0.2 #width of each bar
p = np.arange(len(a)) #provides a sequence of number in an order
q = [j+width for j in p]

plt.xlabel("NAMES", fontsize = 15)
plt.ylabel("CGPA", fontsize = 15)
plt.title("SOU MARKS")


plt.bar(p,b,width, color = "lightblue",label = "2022")
plt.bar(q,c,width, color = "orange",label = "2023")
plt.legend()
plt.xticks()

plt.show()
```

# SCATTER PLOT

```python
day = [1,2,3,4,5,6,7]
num = [2,3,1,4,5,3,6]

colors = [45,60,15,34,54,78,29]
plt.scatter(day, num, marker = "x",c = colors, cmap = "BuGn") #x is chosen as marker in histogram
plt.title("Days v/s number")
plt.xlabel("Day", fontsize = 15) #x-axis label
plt.ylabel("Number", fontsize = 15) #y-axis label

t = plt.colorbar()
t.set_label("color bar") #to label the color bar on the right side of graph
plt.show()
```

# PIE CHART

```python
a = [10,45,30,5,5]
b = ['manthan', 'dev', 'Jani', 'vanshil', 'devasya']

ex = [0,0,0.3,0,0] #explode function used for highlighting the main part by splitting it from pie chart
plt.title("No. of runs in a match")

#labels is used to label the area of pie chart
#autopct is used to mention the percentage automatically, based on the values
plt.pie(a,labels = b, explode = ex, autopct = "%0.1f%%")
plt.show()
```

# STACK PLOT

```python
a1 = [2,4,6,8,10]
area1 = [1,2,3,4,5]
area2 = [2,4,6,10,8]
area3 = [3,6,5,2,12]

l = ['area1','area2', 'area3']
plt.stackplot(a1,area1, area2, area3,labels = l, colors = ["r","g","y"])
plt.legend(loc = 2)
plt.show()
```

# BOX AND WHISKER PLOT 

```python
f=[10, 20, 30, 40, 50, 60,70]
g=[10,20,40,50,30,60,90]
h=[f,g]
plt.boxplot (h, labels = ["python","c++"] ,showmeans =True, sym="g+", boxprops=dict (color="r"), capprops=dict(color="r"), whiskerprops=dict(color="r"))
```
